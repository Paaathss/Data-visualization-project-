**VISUALIZATION**

```python
#Distribution of Preparation Time

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/project/indian_food.csv')


plt.figure(figsize=(9,6))
sns.histplot(df['prep_time'], bins=23,kde=True)
plt.style.available

plt.grid(True,color='pink')
plt.xlabel('Preparation Time (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Preparation Time')
plt.style.use( 'dark_background')

plt.show()
```

**Diet Composition - Veg vs Non Veg**

```python
plt.figure(figsize=(5, 4))

df['diet'].value_counts().plot(kind='bar',color=['beige', 'g'])
plt.xlabel('Diet')
plt.ylabel('Count')
plt.title('Vegetarian vs Non-vegetarian Dishes')

plt.xticks(rotation=0)
plt.show()
```

**Flavour Profile**

```python
plt.figure(figsize=(8,5))
df['flavor_profile'].value_counts().plot(kind='bar', color=['pink', 'beige'],label = 'flavor')
plt.xlabel('Flavor Profile')
plt.ylabel('Count')
plt.title('DISTRIBUTION OF FLAVOR PROFILES')
plt.legend(loc='upper left')


plt.xticks(rotation=80)
plt.show()
```

**Course Distribution**

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(8, 5))
df['course'].value_counts().plot(kind='pie',autopct = '%1.1f%%',shadow = 'True',explode = (0,0,0,1),colors=['beige', 'pink', 'lightblue', 'y'])
plt.title('Distribution of Meal Courses')
plt.ylabel('')
plt.show()
```

**Feature relationships**



```python
sns.scatterplot(x='prep_time', y='cook_time', data=df, color = 'pink')
plt.xlabel('Preparation Time (minutes)')
plt.ylabel('Cooking Time (minutes)')
plt.title('Preparation Time vs. Cooking Time')
plt.show()
```

Box plot

```python
sns.boxplot(df['region'],color = 'y')
```

```python
sns.boxplot(x = 'cook_time',y = 'flavor_profile',data = df,hue = 'diet')
```

```python
# Select only the numeric columns
numeric_columns = ['prep_time', 'cook_time']

# Compute the correlation matrix
corr_matrix = df[numeric_columns].corr()

# Create a heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(corr_matrix, annot=True, cmap='pink', square=True)
plt.title('Correlation Heatmap for Numeric Features')
plt.show()
```

pair plot

```python
sns.pairplot(df,hue = 'flavor_profile')
```

**State Wise cooking time**

```python
plt.figure(figsize=(14, 10))
sns.boxplot(x='state', y='cook_time', data=df)
plt.xlabel('State')
plt.ylabel('Preparation Time (minutes)')
plt.title('Distribution of Preparation Times by State')
plt.xticks(rotation=45)
plt.show()
```
